```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('hacktrain.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(numeric_only=True),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['ID'],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
df['class']=encoder.fit_transform(df['class'])
```

```python
X = df.drop(columns=['class'])
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=40)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'penalty':['l2'],'solver':['lgbs','newton-cg'],'max_iter':[1000],'tol':[1e-4]}
```

```python
clf_reg = GridSearchCV(clf,param_grid=parameters,scoring='accuracy',cv=10)
clf_reg.fit(X_train,y_train)
```

```python
print(clf_reg.best_params_)
```

```python
print(clf_reg.best_score_)
```

```python
y_pred = clf_reg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
print(score)
```

```python
print(classification_report(y_pred,y_test))
```
